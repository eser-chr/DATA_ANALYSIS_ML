```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
from ucimlrepo import fetch_ucirepo 
  
credit_approval = fetch_ucirepo(id=27)  # Code provided by the source No clue what id means
  
X = credit_approval.data.features 
y = credit_approval.data.targets 
```

The attributes name has been substituted by the creator of the dataset with meaningless symbols. This does not allow us to preprocess the data because of their nature. Moreover, the size is relatively small, which mean that we can be generous with the use of RAM and CPU!!!
For those reason the strategy is to apply OneHotEncoding to all categorical attributes (NaN will also be a category by itself) then spilt to train and test and later apply Standardization to the continous attributes.

```python
# OneHotEncoding to all categorical attributes
encoded_X = pd.get_dummies(X, columns=['A13', 'A12','A11','A10','A9','A7','A6','A5','A4','A1'])
```

```python
encoded_X
```

```python
y.loc[:,'A16'] = y['A16'].apply(lambda x: 1 if x=='+' else 0) # Fixing the target symbol to 1 and 0. We had no missing values. So we can do this before trainn test split
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(encoded_X,y) #default size = 25%
```

```python
X_train
```

```python
# Standardization
scaler = StandardScaler()
scaler.fit(X_train[['A15', 'A14', 'A8', 'A3', 'A2']])
```

```python
X_train[['A15', 'A14', 'A8', 'A3', 'A2']]=scaler.transform(X_train[['A15', 'A14', 'A8', 'A3', 'A2']])
X_test[['A15', 'A14', 'A8', 'A3', 'A2']]=scaler.transform(X_test[['A15', 'A14', 'A8', 'A3', 'A2']])
```
